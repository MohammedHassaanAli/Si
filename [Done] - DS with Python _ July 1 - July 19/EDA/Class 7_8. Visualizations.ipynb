# Data Visualizations
- univariate visualizations for a continuous var: histogram, kde, boxplot, violinplot, lineplot(to show how the variance in the data)
- univariate visualizations for a factor var: barplot, pie chart
- bivariate - continuous: scatterplot. kde
- bivariate - factor+continuous: find the mean age according to pclass and plot using barplots
    - you can seperate out the distribution of age according to each level and do a univariate analysis on each of the distributions for comparison purpose
- bivariate - factor + factor: factorplots, sns barplots -%age of survived based on gender
- multivariate: cont+cont+factor: scatter plots and hue according to the factor var
- multivariate: factor+factor+cont: violin plots according to the level in factor variable1 and split color according to factor var2; 2nd we did barplots after grouping according to 2 factor variables find out the mean age and create bar plot
