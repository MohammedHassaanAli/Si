```python
np.random.seed(1) #to replicate the same random numbers multiple times
ran = np.random.rand(100)
ran[:3]
```

```python
plt.figure(figsize = (10,4))
plt.plot(ran, color = '#00FF00', ls = "-.", label = "line1", marker = "o", linewidth = 2, alpha = .9)
#alpha - transparency level - alpha = 1 means 100% opaque line
#color - any hexadecimal code or abbreviations: 'r', 'b', 'g'
#marker - + . 1234 o
#linestyle - ls = (-, --, -., steps)
plt.show()
```

```python
monday = [12,10, 20, 32]
tuesday = [20,39,18, 23]
wednesday=[ 14,34, 23, 27]
time_hrs = [1,2,3,4]
```

```python
plt.figure(figsize = (10,3))
plt.plot(time_hrs,monday, 'r', label = "monday", linewidth = 2)#label important for legend
plt.plot(time_hrs,tuesday, 'g', label = "tuesday", linewidth = 1)
plt.plot(time_hrs,wednesday, 'b', label = "tuesday", linewidth = 1)
plt.legend()
plt.title("Website Traffic")
plt.show()
```

```python
bosdf = pd.read_csv(r"C:\Users\Vaibhav\Desktop\BA\Datasets\boston.csv")
bosdf.head(3)
```

```python
#line plots used to analyse the covariance - covariance means how the variable is 
#varying wrt other variables 
plt.figure(figsize = (10,3))
plt.plot(bosdf["MEDV"],label = "MEDV")#row number in x axis and y = Median value
plt.plot(bosdf['CRIM'], label = "CRIM")#row number in x and y axis - crime rate
plt.plot(bosdf['AGE'], label = "Age")#row number in x axis and y axis has age
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplots_adjust(left = .2, wspace = .6)

plt.subplot(1,2,1) #1 row 2 columns, 1st subplot
plt.plot(monday, "-o", alpha = .1, color = 'r')
plt.title("Monday")

plt.subplot(1,2, 2) #2nd subplot
plt.plot(tuesday, "-", color = "b")
plt.title("Tuesday")
plt.show()
```

### BIVARIATE ANALYSIS - FACTOR + CONTINUOUS VARIABLES

```python
plt.figure(figsize=(10,3))
bosdf.groupby("CHAS").MEDV.mean().plot.bar()
plt.show()
```

```python
sns.violinplot(x = "CHAS", y = "MEDV", data = bosdf)
```

```python
#another way of creating subplots
f,ax = plt.subplots(1,2, figsize = (18,5))
bosdf[bosdf.CHAS == 0].MEDV.plot.hist(ax = ax[0], bins = 10, color = "red")
ax[0].set_title("CHAS = 0")
ax[0].set_xticks(range(5,50,3))
ax[0].set_xlabel("MEDV")

bosdf[bosdf.CHAS ==1].MEDV.plot.hist(ax = ax[1], color = "b")
ax[1].set_title("CHAS=1")
ax[1].set_xticks(range(10,50, 2))
ax[1].set_xlabel("MEDV")
plt.show()
```
