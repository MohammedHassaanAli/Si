```python
boston = np.concatenate((bs.data, y), axis = 1)
boston.shape
import pandas as pd
boston = pd.DataFrame(boston)
print(boston.head(3))
boston.columns = list(bs.feature_names)+['MEDV']
boston.columns
```

```python
X = boston.drop('MEDV', axis=1).values
y = boston['MEDV'].values
print(y)
y = y.reshape(-1, 1) #this means no. of rows is unknown and no. of cols is 1
```
