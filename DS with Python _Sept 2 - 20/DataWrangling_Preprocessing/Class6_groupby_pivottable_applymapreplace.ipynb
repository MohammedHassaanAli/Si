```python
def add_custom_values(x, *args, **kwargs):
    #*args means, any number of arguments
    #**kwargs means that kwargs is a dictionary of argument as the key and its values 
    for k in kwargs:
        x += kwargs[k] #same as x = x+ kwargs[x]
    return x
```

```python
add_custom_values(5, a = 10, b = 20)
```

```python
add_custom_values("Samridhi", a = "Dutta", b = "x", c = "a")
```

```python
brics
```
